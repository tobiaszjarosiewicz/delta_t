```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('prog_bfr.dat', 'r') as f:
     data = f.readlines()
```

```python
# Plot dimmensions:
p_width = 16
p_heigth = 8
```

```python
plot_t = []
plot_bd = []
plot_dtot = []
for i, line in enumerate(data):
    values = line.split()
    plot_t.append(values[0])
    plot_bd.append(values[1])
    plot_dtot.append(values[2])
```

```python
plt.rcParams['figure.figsize'] = [p_width, p_heigth]
plt.plot(plot_t, plot_bd, label='bf distance', marker='o', linewidth=0.2)
plt.plot(plot_t, plot_dtot, label='total distance', marker='o', linewidth=0.2)
plt.xlabel('[t]')
plt.ylabel('[km]')
plt.legend()
plt.grid(True)
plt.show()
```
